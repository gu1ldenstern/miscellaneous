```python
def get_stats(data, s, feature, name):
    n = data
    # n[feature].fillna(0, inplace = True)
    print("Число пропусков: ", n[feature].isna().sum())
    n[feature].dropna(inplace = True)
    n[feature].replace({'Y':1, 'N':0}, inplace = True)
    s[name +' всего наблюдений'] = n.groupby(by = ['region'])[feature].count()
    s[name +' число болеющих'] = n.groupby(by = ['region'])[feature].sum()
    s[name +' % заболевания'] = n.groupby(by = ['region'])[feature].mean() *100
    s.iloc[0, -3:] = [n[feature].count(), n[feature].sum(), n[feature].mean()*100]
    return s
```

```python
def get_op_stats(data, s, feature, name):
    
    s = s.join(pd.DataFrame(np.random.rand(7,9),
                        columns=pd.MultiIndex.from_product([[name], ['все типы операций', 'лапороскопия', 'лапоротомия'], 
                                                           ['всего наблюдений', 'число операций', '% операций']]),
                        index=ops.index))
    n = data.copy()
    print("Число пропусков: ", n[feature].isna().sum())
    n[feature].dropna(inplace = True)
    # все
    all_ops = n.copy()
    all_ops[feature].replace({2:1}, inplace = True)
    s[(name, "все типы операций", 'всего наблюдений')] = all_ops.groupby(by = ['region'])[feature].count()
    s[(name, "все типы операций", 'число операций')] = all_ops.groupby(by = ['region'])[feature].sum()
    s[(name, "все типы операций", '% операций')] = all_ops.groupby(by = ['region'])[feature].mean() *100
    s.iloc[0, -9:-6] = [all_ops[feature].count(), all_ops[feature].sum(), all_ops[feature].mean()*100]
    # лапороскопия
    l2 = n.copy()
    l2[feature].replace({2:0}, inplace = True)
    s[(name, "лапороскопия", 'всего наблюдений')] = l2.groupby(by = ['region'])[feature].count()
    s[(name, "лапороскопия", 'число операций')] = l2.groupby(by = ['region'])[feature].sum()
    s[(name, "лапороскопия", '% операций')] = l2.groupby(by = ['region'])[feature].mean() *100
    s.iloc[0, -6:-3] = [l2[feature].count(), l2[feature].sum(), l2[feature].mean()*100]
    # лапоротомия
    l1 = n.copy()
    l1[feature].replace({1:0, 2:1}, inplace = True)
    s[(name, "лапоротомия", 'всего наблюдений')] = l1.groupby(by = ['region'])[feature].count()
    s[(name, "лапоротомия", 'число операций')] = l1.groupby(by = ['region'])[feature].sum()
    s[(name, "лапоротомия", '% операций')] = l1.groupby(by = ['region'])[feature].mean() *100
    s.iloc[0, -3:] = [l1[feature].count(), l1[feature].sum(), l1[feature].mean()*100]
    return s
```
